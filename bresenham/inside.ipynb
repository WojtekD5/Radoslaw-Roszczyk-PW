```python
def punkt_wewnatrz_prostokata(x1, y1, x2, y2, pX, pY):
  return pX < x2 and pX > x1 and pY < y2 and pY > y1
```

```python
def punkt_na_krawedzi_prostokata(x1, y1, x2, y2, pX, pY):
  return pX == x2 or pX == x1 or pY == y2 or pY == y1
```

```python
def punkt_wewnatrz_okregu(x0, y0, r, pX, pY):
  import sys

  epsilon = sys.float_info.epsilon

  return ((pX - x0) ** 2) + ((pY - y0) ** 2) + epsilon < r ** 2
```

Na poniższym rysunku półprosta o początku A przecina obwód wielokąta w siedmiu punktach, a więc punkt A leży w wielokącie, a półprosta o początku B ma dwa punkty wspólne z obwodem wielokąta, czyli punkt B leży na zewnątrz.
