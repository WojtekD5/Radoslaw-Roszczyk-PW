```python
# instalujemy pakiet który będzie nam potrzebny
!pip3 install ColabTurtle

# importujemy biblioteki
import turtle
import random
import matplotlib.pyplot as plt
import math
from ColabTurtle.Turtle import *
```

```python
# żółw dla Colab'a nie potrafi rysować okręgów, dlatego dodajemy tę funkcje
def circle(radius, extent = 360, steps = 60):
    if extent < 360 and steps == 360:
        steps = extent
    
    theta = extent / steps
    step_size = 2 * radius * math.sin(math.radians(theta / 2))
    left(theta / 2)
    forward(step_size)
    for i in range(1, steps):
        left(theta)
        forward(step_size)
    
    left(theta / 2)        
```

```python
# inicjalizujemy bilbiotekę Żółwia
initializeTurtle()
speed(13)
hideturtle()
bgcolor('white')
color('black')
up()
setpos(550, 400)
down()
fd(300)
left(90)
fd(300)
left(90)
fd(300)
left(90)
fd(300)
left(90)

up()
setposition(550, 250)
down()
circle(150)
```

```python
# liczniki punktów wewnątrz i na zewnątrz okręgu
in_circle = 0
out_circle = 0

# tablica do przechowywania wartości liczby pi
pi_values = []
```

```python
for i in range(5):
  for j in range(1000):
    x = random.randrange(-150, 150)
    y = random.randrange(-150, 150)

    if (x**2 + y**2 > 150**2):
      out_circle = out_circle + 1      
      if (j < 50):
        color('black')
        up()
        goto(x + 400, y + 250)
        down()
        fd(1)
    else:
      in_circle = in_circle + 1      
      if (j < 50):      
        color('red')
        up()
        goto(x + 400, y + 250)
        down()
        fd(1)

    pi = 4.0 * in_circle / (in_circle + out_circle)

    pi_values.append(pi)

    avg_pi_errors = [abs(math.pi - pi) for pi in pi_values]

  # wypisz wartość liczby pi po każdym 1000 iteracji
  print(pi_values[-1])
```

```python
plt.axhline(y = math.pi, color = 'g', linestyle = '-')
plt.plot(pi_values)
plt.xlabel("Powtórzenia")
plt.ylabel("Obliczona wartość Pi")
plt.show()

plt.axhline(y = 0.0, color='g', linestyle = '-')
plt.plot(avg_pi_errors)
plt.xlabel("Powtórzenia")
plt.ylabel("Błąd")
plt.show()
```
