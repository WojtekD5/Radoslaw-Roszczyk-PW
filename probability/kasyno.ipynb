```python
import random
import matplotlib.pyplot as plt
```

```python
wybor = input("Wybierasz parzyste czy nieparzyste? ")

if wybor == "parzyste":
  print("Wybrałeś parzyste - definiuję funkcję")
  def pickNote():
    note = random.randint(1, 100)
    if note % 2 != 0 or note == 10:
      return False
    elif note % 2 == 0:
      return True
elif wybor == "nieparzyste":
  print("Wybrałeś nieparzyste - definiuję funkcję")  
  def pickNote():
    note = random.randint(1, 100)
    if note % 2 == 0 or note == 11:
      return False
    elif note % 2 == 1:
      return True  
```

```python
def graj(kwota_pieniedzy, wartosc_zakladu, liczba_gier):
  num_of_plays = []
  money = []

  for play in range(liczba_gier):
    if pickNote():
      kwota_pieniedzy = kwota_pieniedzy + wartosc_zakladu
    else:
      kwota_pieniedzy = kwota_pieniedzy - wartosc_zakladu
    num_of_plays.append(play)
    money.append(kwota_pieniedzy)      

  plt.ylabel("Pieniądze gracza")
  plt.xlabel("Liczba zakładów")
  plt.plot(num_of_plays, money)

  return(money[-1])
```

```python
wygrane = []

kwota_startowa = 10000

for i in range(100):
  wygrana = graj(kwota_startowa, 100, 50)
  wygrane.append(wygrana)

print("Maksymalna wygrana: ", max(wygrane), " Minimalna wygrana: ", min(wygrane))
print("Gracz rozpoczął grę z kwotą: ", kwota_startowa)
print("Gracz zankończył grę z kwotą: ", sum(wygrane) / len(wygrane))
```
